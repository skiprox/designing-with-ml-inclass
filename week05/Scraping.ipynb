# Scraping

using [this stuff](https://github.com/psf/requests-html)

```python
from requests_html import HTMLSession
session = HTMLSession()
responses = []
base_url = 'https://newyork.craigslist.org/d/free-stuff/search/zip'
responses = [session.get(base_url), session.get(base_url + '?s=120'), session.get(base_url + '?s=240')]
print(responses)
```

```python
link_tags_array = [response.html.find('a.result-title') for response in responses]
# flatten the array of arrays
link_tags = [item for sublist in link_tags_array for item in sublist]

# Get the links from link_tags
links = []
for link in link_tags:
    links.append(link.attrs.get('href'))

print(links)
```

```python
import pandas as pd

clist_df = pd.DataFrame(columns=['id', 'title', 'description', 'images'])
clist_df
```

```python
for link in links:
    print(link)
    try:
        # store response
        link_response = session.get(link)
        # page id
        page_id = link.split('/')[-1].replace('.html', '')
        # page title
        title = link_response.html.find('#titletextonly', first=True).text
        # page description
        description = link_response.html.find('#postingbody', first=True).text.replace('QR Code Link to This Post\n', '')
        # images
        images = [img.attrs.get('src') for img in link_response.html.find('.slide img')] 
        clist_df = clist_df.append({
            'id': page_id,
            'title': title,
            'description': description,
            'images': images
        }, ignore_index=True)
        # do this to not hammer their servers,
        # you can use random number to make you seem less like a machine
        sleep(0.1)
    except:
        print('link broken')

```

```python
import urllib.request
from time import sleep

for index, row in clist_df.iterrows():
    for url in row['images']:
        print(url)
        file_name = url.split('/')[-1]
        urllib.request.urlretrieve(url, 'scraped_images/' + file_name)
        sleep(0.1)
```
